```python
import random
import csv
from datetime import datetime, timedelta
import os

os.makedirs("/Volumes/workspace/default/ecommercepipeline/data", exist_ok=True)


customers_file = "/Volumes/workspace/default/ecommercepipeline/data/customers.csv"
orders_file = "/Volumes/workspace/default/ecommercepipeline/data/orders.csv"
payments_file = "/Volumes/workspace/default/ecommercepipeline/data/payments.csv"

cities = ["Delhi", "Mumbai", "Lucknow", "Bangalore", "Hyderabad", "Pune", "Chennai"]
categories = ["Electronics", "Fashion", "Grocery", "Furniture", "Beauty", "Sports"]
payment_modes = ["UPI", "CARD", "COD", "NETBANKING"]
statuses = ["SUCCESS", "FAILED", "PENDING"]

num_customers = 10000
num_orders = 100000



# customers.csv
with open(customers_file, "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["customer_id", "customer_name", "city", "signup_date", "is_active"])

    for i in range(1, num_customers + 1):
        signup_date = datetime(2022, 1, 1) + timedelta(days=random.randint(0, 1000))
        writer.writerow([
            i,
            f"Customer_{i}",
            random.choice(cities),
            signup_date.date(),
            random.choice([True, False])
        ])

# orders.csv
with open(orders_file, "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow([
        "order_id",
        "customer_id",
        "order_date",
        "category",
        "quantity",
        "unit_price",
        "discount",
        "order_status"
    ])

    for i in range(1, num_orders + 1):
        order_date = datetime(2023, 1, 1) + timedelta(days=random.randint(0, 700))
        quantity = random.randint(1, 5)
        unit_price = round(random.uniform(100, 5000), 2)
        discount = random.choice([0, 5, 10, 15, 20, None])

        writer.writerow([
            i,
            random.randint(1, num_customers),
            order_date.date(),
            random.choice(categories),
            quantity,
            unit_price,
            discount,
            random.choice(["DELIVERED", "CANCELLED", "RETURNED"])
        ])

# payments.csv
with open(payments_file, "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["payment_id", "order_id", "payment_mode", "payment_status", "payment_date"])

    for i in range(1, num_orders + 1):
        payment_date = datetime(2023, 1, 1) + timedelta(days=random.randint(0, 700))
        writer.writerow([
            i,
            i,
            random.choice(payment_modes),
            random.choice(statuses),
            payment_date.date()
        ])

print("Dataset generated successfully.")
```
